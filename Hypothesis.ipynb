What is Estimation Statistics? Explain point estimate and interval estimate.


##### Ans
Estimation statistics involves using sample data to estimate unknown population parameters.  
Point Estimate: A single value calculated from the sample data, used as an estimate of the population parameter. For example, the sample mean is a point estimate of the population mean.

Interval Estimate: Provides a range of values within which the population parameter is likely to be found. Confidence intervals are a common type of interval estimate, indicating a range of values with a specified level of confidence.

#### Number 7

##### Ans
The steps involved in hypothesis testing are:

1.Formulate the null hypothesis (H0) and alternative hypothesis (Ha).  
2.Determine the appropriate test statistic and its sampling distribution.  
3.Specify the significance level (alpha) to define the threshold for rejecting the null hypothesis.  
4.Collect sample data and calculate the test statistic.    
5.Determine the p-value associated with the test statistic.  
6.Compare the p-value with the significance level to make a decision.  
7.Interpret the results and draw conclusions.

#### Number 8
Define p-value and explain its significance in hypothesis testing.

##### Ans
The p-value is the probability of obtaining the observed data, or data more extreme, assuming that the null hypothesis is true. It measures the strength of evidence against the null hypothesis. A p-value below the significance level (alpha) indicates that the observed data is unlikely to have occurred by chance alone, leading to the rejection of the null hypothesis.

#### Number 9
Write code in Python to plot t-Distribution, with degrees of freedom as 10

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

df = 10  # Degrees of freedom

x = np.linspace(-4, 4, 100)
y = t.pdf(x, df)

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title("Student's t-Distribution (df=10)")
plt.grid(True)
plt.show()
```

#### Number 13
A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random
sample of 50 days and find the sample mean revenue to be R.500 with a standard deviation of R.50.
Estimate the population mean revenue with a 95% confidence interval.

```python
from scipy import stats
sample_mean = 500
sample_std_dev = 50
sample_size = 50
confidence_level = 0.95
alpha = 1 - confidence_level

standard_error = sample_std_dev / (sample_size ** 0.5)
critical_value = stats.t.ppf(1 - alpha / 2, sample_size - 1)
margin_of_error = critical_value * standard_error
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Lower bound is {lower_bound} and Upper Bound is {upper_bound}")
```

#### Number 14
A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

```python
import numpy as np
from scipy import stats

n = 100
pop_mean = 10
sample_mean = 8
std_dev = 3
alpha = 0.05

# Generate the random sample
sample_data = np.random.normal(loc=pop_mean, scale=std_dev, size=n)

# Perform the one-sample t-test
t_stat, p_value = stats.ttest_1samp(a=sample_data, popmean=pop_mean)

# Print the results
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Check if the null hypothesis is rejected based on the significance level
if p_value < alpha:
    print("Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")
```

#### number 16
Two groups of students are given different study materials to prepare for a test. The first group (n1 =
30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
groups are equal with a significance level of 0.01.

```python
import numpy as np
import scipy.stats as stats

n1,n2=30,40
mean1,mean2=80,75
std_dev1,std_dev2=10,8
t_statistic,p_value=stats.ttest_ind_from_stats(mean1=mean1,mean2=mean2,std1=std_dev1,std2=std_dev2,nobs1=n1-1,nobs2=n2-1,equal_var=False)
alpha = 0.01

if p_value < alpha:
    print("Reject the null hypothesis. The population means for the two groups are different.")
else:
    print("Fail to reject the null hypothesis. The population means for the two groups are equal.")
```
